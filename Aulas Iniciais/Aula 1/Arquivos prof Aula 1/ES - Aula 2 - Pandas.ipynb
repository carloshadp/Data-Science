```python
df_filtered = df[df['Status_Pagamento'] == 'Pendente']
def destacar_pendente(valor):
    return 'font-weight: bold; color: orange' if valor == 'Pendente' else ''

df_filtered.style.applymap(destacar_pendente, subset=['Status_Pagamento'])
```
