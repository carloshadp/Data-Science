```python
import pandas as pd
```

```python
#1
df = pd.read_csv("A://FIAP//2 Sem//MATÉRIAS//Data-Science//Atividade Employees//employees.csv")
df.head()
```

```python
#2
TI = df[df["Departamento"] == "TI"]
TI
```

```python
#3
salarios = df["Salário"]
soma = 0
media = None
for salario in salarios:
    soma+=salario
media = soma / len(salarios)
print(f"A média dos salários é: ",media)
```

```python
#4
admiticao_arrumado = []
splited = None
Admitidos_2020 = None
for data in df["Data_Admissão"]:
    splited = data.split("-")
    admiticao_arrumado.append(int(splited[0]))
    
df["Adimição_Arrumado"] = admiticao_arrumado

Admitidos_2020 = df[df["Adimição_Arrumado"] > 2019]
Admitidos_2020
```

```python
#5
Anos_na_Empresa = []
Tempo = None
for data in df["Adimição_Arrumado"]:
    Tempo = 2025 - int(data)
    Anos_na_Empresa.append(Tempo)
df["Tempo_Empresa"] = Anos_na_Empresa
df.head()
```

```python
#6
novo = df.iloc[0]
velho = df.iloc[0]

for i in range(len(df["Idade"])):
    if df.iloc[i]["Idade"] > velho["Idade"]:
        velho = df.iloc[i]
    if df.iloc[i]["Idade"] < novo["Idade"]:
        novo = df.iloc[i]
        
print(novo, "\n", velho)
```

```python
#7
media_salario_cargo = df.groupby("Cargo")["Salário"].mean()
print(media_salario_cargo.idxmax(), media_salario_cargo.max())
```

```python
#8
media_sal_departamento = df.groupby("Departamento")["Salário"].mean()
media_sal_departamento
```

```python
#9
df = df.sort_values("Tempo_Empresa", ascending=False)
df["Tempo_Empresa"]
```
