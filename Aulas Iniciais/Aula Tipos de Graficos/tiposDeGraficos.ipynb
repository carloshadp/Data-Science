TIPOS DE GRÁFICOS

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#Gráfico de Linha Simples

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Meu Primeiro Gráfico')
plt.show()
```

```python
#Scatter / Dispersão

x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color="blue")
plt.xlabel("Variavel X")
plt.ylabel("Variavel Y")
plt.title("Exemplo de Scatter Plot")
plt.show()
```

```python
#Gráfico de Barras
categorias = ['A', 'B', 'C', 'D']
valores = [4, 7, 1, 8]

plt.bar(categorias, valores)
plt.show()
```

```python
#HISTOGRAMA
dados = np.random.randn(1000)

plt.hist(dados, bins=20, edgecolor="red")
plt.show()
```

```python
# BOXPLOT
dados = [7,8,9,10,11,12,50]
plt.boxplot(dados)
plt.show()
```

```python
#HEATMAP

dias = ["Seg", "Ter", "Qua", "Qui", "Sex", "Sáb", "Dom"]
horas = ["Manhã", "Tarde", "Noite", "Madrugada", "Almoço", "Happy Hour"]
vendas = np.random.randint(0, 50, (7, 6))  

plt.figure(figsize=(8, 5))
sns.heatmap(vendas, annot=True, cmap="coolwarm", xticklabels=horas, yticklabels=dias)

plt.title("Heatmap de Vendas por Dia e Período")
plt.xlabel("Período do Dia")
plt.ylabel("Dia da Semana")
plt.show()
```

```python
#PERSONALIZANDO GRÁFICOS
x= [1, 2, 3, 4, 5]
y= [10, 20, 30, 40, 50]

plt.plot(x, y, color="red", linestyle="--", marker="s", label="Série de Dados")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.legend()
plt.grid()
plt.show()
```
